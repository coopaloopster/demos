```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

### Triangular Edge

```python
def f(s,x):
    cos = 0
    sin = 0
    for i in range(s+1):
        cos += np.cos((2*i+1)*x)/(2*i+1)**2
    for i in range(1,s+1):
        sin = sin + np.sin(i*x)/i
    return np.pi/4+2/np.pi*cos+sin
```

```python
x = np.linspace(-np.pi,np.pi,300)
plt.plot(x,f(3,x),x,f(30,x),x,f(300,x))
```

### Saw Tooth

```python
def g(s,x):
    sin = 0
    for i in range(1,s+1):
        sin = sin + ((-1)**(i+1))*np.sin(i*x)/i
    return 2*sin
```

```python
x = np.linspace(-2*np.pi,2*np.pi,300)
plt.plot(x,g(3,x),x,g(30,x),x,g(300,x))
```

### Triangle Wave

```python
def h(s,x):
    cos = 0
    for i in range(s+1):
        cos = cos + np.cos((2*i+1)*np.pi*x)/(2*i+1)**2
    return (8/np.pi**2)*cos
```

```python
x = np.linspace(-np.pi,np.pi,300)
plt.plot(x,h(100,x))
```

### Square Wave

```python
def l(s,x):
    sin = 0
    for i in range(s+1):
        sin = sin + np.sin((2*i+1)*np.pi*x)/(2*i+1)
    return (4/np.pi)*sin
```

```python
x = np.linspace(-np.pi,np.pi,300)
plt.plot(x,l(1,x),x,l(3,x),x,l(300,x))
```

### Gibb's Phenomenon

```python
x = np.linspace(0,.3,1000)
plt.plot(x,l(100,x))
plt.xlim(0,.3)
plt.ylim(.5,1.5)
```

### Even and Odd Extensions

```python
def even(s,x):
    cos = 0
    for n in range(1,s):
        cos += ((-1)**n*np.e - 1)/(1 + n**2*np.pi**2)*np.cos(n*np.pi*x)
    return (np.e - 1) + 2*cos
```

```python
x = np.linspace(-2,2,300)
plt.plot(x,even(2,x),x,even(3,x),x,even(5,x),x,even(50,x))
```

```python
def odd(s,x):
    sin = 0
    for n in range(1,s):
        sin += n*(1 - (-1)**n*np.e)/(1 + n**2*np.pi**2)*np.sin(n*np.pi*x)
    return 2*np.pi*sin
```

```python
x = np.linspace(-2,2,300)
plt.plot(x,odd(2,x),x,odd(3,x),x,odd(5,x),x,odd(50,x))
```

### All together

```python
def e(s,x):
    sums = 0
    for n in range(1,s+1):
        sums += (((-1)**n)/(1 + n**2*np.pi**2))*(np.cos(n*np.pi*x) - n*np.pi*np.sin(n*np.pi*x))
    return (np.e - 1/np.e)*(.5 + sums)
```

```python
x = np.linspace(-2,2,300)
plt.plot(x,e(2,x),x,e(3,x),x,e(5,x),x,e(50,x))
```
